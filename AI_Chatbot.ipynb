```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

# Configure 4-bit quantization for efficient memory usage
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

# Load the model and tokenizer
model_name = "ruslanmv/ai-medical-model-32bit"
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    trust_remote_code=True,
    use_cache=False,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
```

```python
def askme(question):
    # Create the prompt for the model
    prompt = f"<|start_header_id|>system<|end_header_id|> You are a Medical AI chatbot assistant. <|eot_id|><|start_header_id|>User: <|end_header_id|>This is the question: {question}<|eot_id|>"

    # Tokenize the input and generate the output
    inputs = tokenizer([prompt], return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=256, use_cache=True)
    answer = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]

    # Clean up the answer by removing the prompt
    try:
        answer_parts = answer.split("\n", 1)
        if len(answer_parts) > 1:
            return answer_parts[1].strip()
        else:
            return answer
    except:
        return answer
```

```python
# Example usage
question = "What is the cause of Alopecia? How to treat it?"
response = askme(question)
print("Healthcare Assistant:", response)
```

```python
# Step 4: Save Streamlit App to File
app_code = """
import streamlit as st
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

# Configure 4-bit quantization for efficient memory usage
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

# Load the model and tokenizer
model_name = "ruslanmv/ai-medical-model-32bit"
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    trust_remote_code=True,
    use_cache=False,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

def askme(question):
    # Create the prompt for the model
    prompt = f"<|start_header_id|>system<|end_header_id|> You are a Medical AI chatbot assistant. <|eot_id|><|start_header_id|>User: <|end_header_id|>This is the question: {question}<|eot_id|>"

    # Tokenize the input and generate the output
    inputs = tokenizer([prompt], return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=256, use_cache=True)
    answer = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]

    # Clean up the answer by removing the prompt
    try:
        answer_parts = answer.split("\\n", 1)  # ✅ Fixed possible escaping issue
        if len(answer_parts) > 1:
            return answer_parts[1].strip()
        else:
            return answer
    except:
        return answer

# Streamlit UI
def main():
    st.title("Healthcare Assistant Chatbot 🩺")
    st.write("Welcome! I'm your AI-powered healthcare assistant. How can I help you today?")

    user_input = st.text_input("Ask your medical question here:", "")

    if st.button("Submit"):
        if user_input:
            st.write("**You:**", user_input)
            with st.spinner("Generating response..."):
                response = askme(user_input)
            st.write("**Healthcare Assistant:**", response)
        else:
            st.warning("Please enter a question.")

if __name__ == '__main__':
    main()
"""

with open("app.py", "w",encoding="utf-8") as f:
    f.write(app_code)
```
